```python
import lxml
Name=[]
#Age=[]
Title=[]
Followers=[]
Connections=[]
engagement=[]
eng_3_post=[]
posts_in_weeks=[]
last_post=[]
time_post=['1m', '2m', '3m', '4m', '5m', '6m', '7m', '8m', '9m', '10m', '11m', '12m', '13m', '14m', '15m', '16m', '17m', '18m', '19m', '20m', '21m', '22m', '23m', '24m', '25m', '26m', '27m', '28m', '29m', '30m', '31m', '32m', '33m', '34m', '35m', '36m', '37m', '38m', '39m', '40m', '41m', '42m', '43m', '44m', '45m', '46m', '47m', '48m', '49m', '50m', '51m', '52m', '53m', '54m', '55m', '56m', '57m', '58m', '59m',
      '1h', '2h', '3h', '4h', '5h', '6h', '7h', '8h', '9h', '10h', '11h', '12h', '13h', '14h', '15h', '16h', '17h', '18h', '19h', '20h', '21h', '22h', '23h', '24h',
      '1d','2d','3d','4d','5d','6d',
      '1w'
     ]
c=0
for i in links:
    c=c+1
    #Open Link
    browser.get(i)
    random_number = random.randint(4, 10)
    time.sleep(random_number)

    # Scrolling to get full page loaded
    SCROLL_PAUSE_TIME = 5
    # Get scroll height
    last_height = browser.execute_script("return document.body.scrollHeight")

    for i in range(3):
        # Scroll down to bottom
        browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait to load page
        time.sleep(SCROLL_PAUSE_TIME)

        # Calculate new scroll height and compare with last scroll height
        new_height = browser.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
    browser.execute_script("window.scrollBy(0,-20000);")
    #Getting HTMML tags
    src = browser.page_source
    soup = BeautifulSoup(src, 'html.parser')
    experience = soup.find_all("section", {"class": "artdeco-card ember-view relative break-words pb3 mt2"})

    #Scrapping name
    name=browser.find_element(By.TAG_NAME,'h1').text.strip('\n...')
    name

    #Scrapping bio
    try:
        title = soup.find("div", {'class': 'text-body-medium break-words'})
        bio = title.get_text().strip()
    except:
        bio='None'

    #Followers
    try:
        followers = soup.find("li", {'class': 'text-body-small t-black--light inline-block'})
        followers=followers.find('span')
        no_of_fol = followers.get_text().strip()
    except AttributeError:
        for i in range(len(experience)):
            tag_exp=experience[i].find('span').get_text().strip()
            if tag_exp=='Activity':
                break
            else:
                continue
        followers=experience[i].find('p')
        followers=followers.find('span',{'aria-hidden':'true'})
        no_of_fol = followers.get_text().strip().replace(' followers', '')
    except:
        no_of_fol='None'

    #Connection
    try:
        conn = soup.find_all("li", {'class': 'text-body-small'})[1]
        no_of_con = conn.get_text().strip()
        if no_of_con=='500+ connections':
            no_of_con='500+'
    except IndexError:
        conn = soup.find("li", {'class': 'text-body-small'})
        no_of_con = conn.get_text().strip()
        if no_of_con=='500+ connections':
            no_of_con='500+'
    except:
        no_of_con='None'

    #engagement
    try:
        for i in range(len(experience)):
            tag_exp=experience[i].find('span').get_text().strip()
            if tag_exp=='Activity':
                break
            else:
                continue
        react=experience[i].find_all('span',{'class','social-details-social-counts__reactions-count'})
        react_no=[]
        for y in range(3):
            try:
                reacts=int(react[y].get_text().strip().replace(',', ''))
                react_no.append(reacts)
            except:
                reacts=0
                react_no.append(reacts)
        react_total=int(sum(react_no)/len(react_no))
    except:
        react_total=0

    #click on post button
    browser.execute_script("window.scrollBy(0,-12000);")
    try:
        comment_load_button=browser.find_element(By.TAG_NAME,'footer')
        comment_load_button.click()
        time.sleep(4)

        try:
            comment_button = WebDriverWait(browser, 10).until(EC.element_to_be_clickable((By.XPATH, './/span[text()="Posts"]')))
            comment_button.click()
            time.sleep(4)
        except:
            pass
        #Scrolling browser
        browser.execute_script("window.scrollBy(0,200000);")
        time.sleep(3)
        browser.execute_script("window.scrollBy(0,-200000);")
        time.sleep(6)

        #avg reaction and total no. of post in 2 weeks and last posted time
        src2 = browser.page_source
        soup2 = BeautifulSoup(src2, 'lxml')
        ul_tag=soup2.find('ul',{'class':'display-flex flex-wrap list-style-none justify-center'})
        li_tag=ul_tag.find_all('li',{'class':'profile-creator-shared-feed-update__container'})
        time.sleep(3)
        count = 0
        avg_react=[]
        post_date=[]
        for z in range(len(li_tag)):
            try:
                tim=li_tag[z].find('div',{'class':'update-components-text-view break-words'})
                tim=tim.find('span',{'aria-hidden':'true'}).get_text().strip().replace(' •', '')
                post_date.append(tim)
            except AttributeError:
                break
            if tim in time_post:
                count += 1
                try:
                    post=li_tag[z].find('span',{'class':'social-details-social-counts__reactions-count'})
                    post_react=int(post.get_text().strip().replace(',', ''))
                    avg_react.append(post_react)
                except AttributeError:
                    post_react=0
                    avg_react.append(post_react)
        if count==5:
            count='5+'
        average = int(sum(avg_react) / len(avg_react))
    except:
        average=0
        count=0
    #appending for making dataset
    
    Name.append(name)
        #Age.append(age)
        #Exp.append(yr_exp)
    Title.append(bio)
    Followers.append(no_of_fol)
    Connections.append(no_of_con)
    engagement.append(average)
    eng_3_post.append(react_total)
    posts_in_weeks.append(count)
    try:
        last_post.append(post_date[0])
    except:
        last_post.append('-')
        #Des.append(designation)
        #Loca.append(work_loc)
        #Com_Name.append(company_name)
        #Edu.append(recent_edu)
        #Re_Skl.append([related_skills])
    
    random_number = random.randint(3, 10)
    time.sleep(random_number)
    print(f'Link {c} is completed')
    
```
